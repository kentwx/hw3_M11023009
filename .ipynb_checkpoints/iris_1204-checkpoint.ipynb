```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import datetime as DT
from sklearn.metrics import accuracy_score
```

# Purity指標運算

```python
def purity_score(Y,label_pred):    
    y_voted_labels = np.zeros(Y.shape)
    labels = np.unique(Y)
    ordered_labels = np.arange(labels.shape[0])
    
    for k in range(labels.shape[0]):
        Y[Y==labels[k]] = ordered_labels[k]
    labels = np.unique(Y)
    bins = np.concatenate((labels, [np.max(labels)+1]), axis=0)

    for cluster in np.unique(label_pred):
        hist, _ = np.histogram(Y[label_pred==cluster], bins=bins)
        winner = np.argmax(hist)
        y_voted_labels[label_pred==cluster] = winner

    return accuracy_score(Y, y_voted_labels)
```

# 原始資料 (依類別分群)

```python
iris = load_iris()

scatter = plt.scatter(iris.data[:,0],iris.data[:,1],c=iris.target)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.legend(handles=scatter.legend_elements()[0],labels=iris.target_names.tolist())
plt.show()
```

# KMeans

```python
ST_time = DT.datetime.now()

estimator = KMeans(n_clusters=3).fit(iris.data)

scatter = plt.scatter(iris.data[:,0],iris.data[:,1],c=estimator.labels_)
plt.xlabel("sepal length")
plt.ylabel("sepal width")

plt.legend(handles=scatter.legend_elements()[0],labels=iris.target_names.tolist())
#畫出質心
plt.scatter(estimator.cluster_centers_[:,0],estimator.cluster_centers_[:,1],marker='*',c='red',s=100)
plt.show()

ED_time = DT.datetime.now()

print("Time:",ED_time-ST_time)
print("Purity:",purity_score(iris.target, estimator.labels_))
```

# 階層式分群

```python
ST_time = DT.datetime.now()
iris=load_iris()
X=iris.data
X=X[:,0:2]

dis=sch.linkage(X,metric='euclidean',method='ward')
k=3
clusters=sch.fcluster(dis,k,criterion='maxclust')
sch.dendrogram(dis)
plt.title('Hierarchical Clustering')
plt.show()

scatter = plt.scatter(X[:,0],X[:,1],c=clusters)
plt.legend(handles=scatter.legend_elements()[0],labels=iris.target_names.tolist())
plt.xlabel("sepal length")
plt.ylabel("sepal width")
#畫出質心
plt.scatter(estimator.cluster_centers_[:,0],estimator.cluster_centers_[:,1],marker='*',c='red',s=100)
plt.show()

ED_time = DT.datetime.now()

print("Time:",ED_time-ST_time)
print("Purity:",purity_score(iris.target, clusters-1))
```

# DBSCAN

```python
ST_time = DT.datetime.now()

iris=load_iris()
X=iris.data
Y=iris.target
X=X[:, :4]

clustering=DBSCAN(eps=0.4,min_samples=4).fit(X)
label_pred = clustering.labels_
#畫圖
x0 = X[label_pred == 0]
x1 = X[label_pred == 1]
x2 = X[label_pred == 2]
plt.scatter(x0[:, 0], x0[:, 1], c="orange", marker='x', label='setosa')  
plt.scatter(x1[:, 0], x1[:, 1], c="blue", marker='*', label='versicolor')  
plt.scatter(x2[:, 0], x2[:, 1], c="brown", marker='+', label='virginica')  
plt.xlabel('sepal length')  
plt.ylabel('sepal width')  
plt.legend(loc=1) 
#畫出質心
plt.scatter(estimator.cluster_centers_[:,0],estimator.cluster_centers_[:,1],marker='*',c='red',s=100)
plt.show() 

ED_time = DT.datetime.now()

print("Time:",ED_time-ST_time)
print("Purity:",purity_score(Y,label_pred))
```
